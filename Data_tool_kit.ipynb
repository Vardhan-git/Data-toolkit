1. What is NumPy, and why is it widely used in Python?
- NumPy is a Python library for numerical computations using arrays. It provides fast, memory efficient arrays and vectorized operations. It's widely used for scientific, statistical, and engineering applications due to its speed and integration with other libraries.

2. How does broadcasting work in NumPy?
- Broadcasting allows NumPy to perform operations between arrays of different shapes by automatically expanding one or both arrays so their shapes match. This avoids loops and improves efficiency.

3. What is a Pandas DataFrame?
- A Pandas DataFrame is a two-dimensional, labeled data structure that holds data in rows and columns, supporting many data analysis and manipulation operations.

4. Explain the use of the groupby() method in Pandas.
- groupby() is used to split data into groups based on column values, perform computations like sum or mean on each group, and combine results. Useful for aggregation and transformation tasks.

5. Why is Seaborn preferred for statistical visualizations?
- Seaborn provides high-level, visually appealing plots and simplifies complex statistical graphs. It integrates with pandas and has built-in themes and color palettes.

6. What are the differences between NumPy arrays and Python lists?
- NumPy arrays offer fast, memory-efficient operations, fixed data types, and support for vectorized calculations; Python lists are slower, store elements of different types, and don't support mathematical operations natively.

7. What is a heatmap, and when should it be used?
- A heatmap displays data values as colors in a matrix format. Use it to visualize correlations, confusion matrices, or large data tables to highlight patterns.

8. What does the term vectorized operation mean in NumPy?
- Vectorized operations allow operations on entire arrays rather than element by element. They are faster and make code more concise.

9. How does Matplotlib differ from Plotly?
- Matplotlib is static and best for basic plots; Plotly creates interactive, web-based visualizations with more modern aesthetics.

10. What is the significance of hierarchical indexing in Pandas?
- Hierarchical indexing allows multiple index levels in rows/columns, letting you work with higher-dimensional data conveniently in two-dimensional DataFrames.

11. What is the role of Seaborns pairplot() function?
- pairplot() creates a matrix of scatterplots for each pair of variables, useful for visualizing relationships and distributions in multivariate datasets.

12. What is the purpose of the describe() function in Pandas?
- describe() provides summary statistics for numeric columns in a DataFrame.

13. Why is handling missing data important in Pandas?
- Handling missing data avoids errors and inaccurate analysis since missing values can skew results, charts, and statistical computations.

14. What are the benefits of using Plotly for data visualization?
- Plotly supports interactive, zoomable, and shareable visualizations works in web apps handles 3D and complex plots with ease.

15. How does NumPy handle multidimensional arrays?
- NumPy can create arrays of any dimensionality, easily supporting 2D matrices, 3D tensors, etc., with powerful indexing and reshaping.

16. What is the role of Bokeh in data visualization?
- Bokeh creates interactive, web-ready visualizations for modern browsers, particularly suited for real-time streaming or large datasets.

17. Explain the difference between apply() and map() in Pandas.
- map() applies a function to each element in a Series.

- apply() applies a function along an axis in DataFrames or on entire Series.

18. What are some advanced features of NumPy?
- Broadcasting, masked arrays, structured arrays, advanced indexing, linear algebra, random sampling, and FFT support.

19. How does Pandas simplify time series analysis?
- It offers datetime indexing, resampling, frequency conversion, rolling window statistics, and time-zone handling, making time series manipulations easy.

20. What is the role of a pivot table in Pandas?
- pivot_table() summarizes data, reshapes it by aggregating values based on column/row groupings.

21. Why is NumPy’s array slicing faster than Python’s list slicing?
- NumPy slices create views and arrays use contiguous memory, while Python lists are pointers and require copying for slices.

22. What are some common use cases for Seaborn?
- Visualizing distributions, statistical relationships, regression lines, heatmaps for correlations, and categorical data plots.

```python
#Create a 2D NumPy array and calculate the sum of each row
import numpy as np
arr = np.array([[1, 2, 3], [4, 5, 6]])
row_sums = np.sum(arr, axis=1)
print(row_sums)
```

```python
#24. Write a Pandas script to find the mean of a specific column in a DataFrame
import pandas as pd
df = pd.DataFrame({'A': [10, 20, 30], 'B': [5, 15, 25]})
mean_A = df['A'].mean()
print(mean_A)
```

```python
#Create a scatter plot using Matplotlib
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')
plt.show()
```

```python
# Calculate the correlation matrix using Seaborn and visualize with a heatmap
import seaborn as sns
import pandas as pd
data = pd.DataFrame({'A': [1, 2, 3], 'B': [5, 6, 7], 'C': [10, 11, 12]})
corr = data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
#Generate a bar plot using Plotly
import plotly.express as px
df = pd.DataFrame({'x': ['a', 'b', 'c'], 'y': [4, 1, 2]})
fig = px.bar(df, x='x', y='y', title='Bar Plot')
fig.show()
```

```python
#Create a DataFrame and add a new column based on an existing column
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3]})
df['B'] = df['A'] * 10
print(df)
```

```python
#Write a program to perform element-wise multiplication of two NumPy arrays
import numpy as np
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
result = arr1 * arr2
print(result)
```

```python
#Create a line plot with multiple lines using Matplotlib
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [2, 5, 10, 17]
plt.plot(x, y1, label='y1')
plt.plot(x, y2, label='y2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Multiple Lines')
plt.legend()
plt.show()
```

```python
#Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold
import pandas as pd
df = pd.DataFrame({'A': [1, 5, 10, 3]})
filtered_df = df[df['A'] > 4]
print(filtered_df)
```

```python
#Create a histogram using Seaborn to visualize a distribution
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
data = np.random.randn(100)
sns.histplot(data, kde=True)
plt.title('Distribution Histogram')
plt.show()
```

```python
#Perform matrix multiplication using NumPy
import numpy as np
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
result = np.dot(a, b)
print(result)
```
